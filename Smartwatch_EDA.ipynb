```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("Smartwatches - smartwatches.csv")

df["Price (USD)"] = pd.to_numeric(df["Price (USD)"], errors="coerce")

# Check the shape and first few rows of the dataset
print("Dataset shape:", df.shape)
print(df.head())

# Check the data types and missing values
print(df.info())

# Check the distribution of each feature
for column in df.columns:
    if column not in ["Brand", "Model"]:
        plt.figure(figsize=(10,2.5))
        sns.histplot(df[column])
        plt.title(column)
        plt.show()

# Check the correlation between features
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation matrix")
plt.show()

# Check the top brands and models
top_brands = df["Brand"].value_counts().head(10)
print("Top brands:\n", top_brands)

top_models = df["Model"].value_counts().head(10)
print("Top models:\n", top_models)





# Check the average price by each feature
for column in df.columns:
    if column not in ["Brand", "Model", "Price (USD)"]:
        plt.figure(figsize=(10,5))
        sns.barplot(x=column, y="Price (USD)", data=df)
        plt.title("Average price by "+column)
        plt.show()
```
